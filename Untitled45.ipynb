```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Define file path
file_path = "/content/city-populations-to-2035.csv"

# Check if file exists
if not os.path.exists(file_path):
    print("Error: File not found. Please check the file path.")
else:
    # Read the dataset
    df = pd.read_csv(file_path)
    print("Dataset Loaded Successfully!")
```

```python
# Display dataset info (Column names, Data types, Missing values)
print("Dataset Info:")
df.info()
```

```python
# Display shape and number of elements
print(f"Dataset Shape: {df.shape}")
print(f"Total Elements in Dataset: {df.size}")
```

```python
# Display column names
print("Column Names:", df.columns.tolist())
```

```python
# Display the first five rows
df.head()
```

```python
# Rename columns for easier access
df.rename(columns={"City population (UN Urbanization Prospects, 2018)": "City Population"}, inplace=True)

# Drop rows where 'City Population' is missing
df_cleaned = df.dropna(subset=["City Population"])
print(f"Missing values handled. Remaining Rows: {df_cleaned.shape[0]}")
```

```python
# Convert 'Year' column to integer
df_cleaned["Year"] = df_cleaned["Year"].astype(int)

# Display updated data types
df_cleaned.dtypes
```

```python
# Remove duplicate rows
df_cleaned = df_cleaned.drop_duplicates()
print(f"Duplicate rows removed. Dataset now has {df_cleaned.shape[0]} rows.")
```

```python
# Identify and remove outliers using the Interquartile Range (IQR) method
q1 = df_cleaned["City Population"].quantile(0.25)
q3 = df_cleaned["City Population"].quantile(0.75)
iqr = q3 - q1

# Filter out outliers
df_filtered = df_cleaned[(df_cleaned["City Population"] >= q1 - 1.5 * iqr) &
                         (df_cleaned["City Population"] <= q3 + 1.5 * iqr)]

print(f"Outliers removed. Remaining Rows: {df_filtered.shape[0]}")
```

```python
# Select specific cities for visualization
selected_cities = ["New York", "London", "Tokyo", "Mumbai", "Bangalore"]
df_selected = df_filtered[df_filtered["Entity"].isin(selected_cities)]

print(f"Filtered Data for Selected Cities: {df_selected.shape[0]} rows")
```

```python
# Plot city population trends over time
plt.figure(figsize=(10, 6))

for city in selected_cities:
    city_data = df_selected[df_selected["Entity"] == city]
    plt.plot(city_data["Year"], city_data["City Population"], label=city)

# Customize the plot
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("City Population Trends (1950-2035)")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Ensure valid data for the latest available year
latest_year = df["Year"].max()
df_latest = df[df["Year"] == latest_year].dropna(subset=["City Population"])

# If df_latest is still empty, use a previous year with valid data
if df_latest.empty:
    valid_years = df.dropna(subset=["City Population"])["Year"].unique()
    latest_year = max(valid_years)  # Find the last available year with data
    df_latest = df[df["Year"] == latest_year]

plt.figure(figsize=(12, 6))
top_cities = df_latest.nlargest(10, "City Population")
plt.bar(top_cities["Entity"], top_cities["City Population"], color='skyblue')
plt.xlabel("City")
plt.ylabel("Population")
plt.title(f"Top 10 Most Populated Cities in {latest_year}")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df["City Population"], bins=30, color='purple', alpha=0.7)
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.title("Distribution of City Populations")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df["City Population"], color='orange')
plt.title("City Population Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for city in selected_cities:
    city_data = df_selected[df_selected["Entity"] == city]
    plt.scatter(city_data["Year"], city_data["City Population"], label=city)
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("City Population Growth Trends")
plt.legend()
plt.grid()
plt.show()
```
